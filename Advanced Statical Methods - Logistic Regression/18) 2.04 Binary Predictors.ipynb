# Import the relevant libraries

```python
import numpy as np
import pandas as pd
import statsmodels.api as sm
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()

# Apply a fix to the statsmodels library
from scipy import stats
stats.chisqprob = lambda chisq, df: stats.chi2.sf(chisq, df)
```

## Load the data

```python
raw_data = pd.read_csv('2.02. Binary predictors.csv')

# Display data
raw_data
```

```python
# Map the variables

# Copy the data in a new variable
data = raw_data.copy()

# Assign 1 to 'Yes' and 0 to 'No'
data['Admitted'] = data['Admitted'].map({'Yes':1, 'No':0})

# Assig the variable Male = 1 and Female = 0 
data['Gender'] = data['Gender'].map({'Female':1, 'Male':0})

# Check data
data
```

## Declare the dependent and the independent variables

```python
# Y variable will hold the Admitted data
y = data['Admitted']

# X1 variable will hold the Gender data
# x1 = data['Gender']

# Changing data frame to predict SAT and Geder
x1 = data[['SAT', 'Gender']]
```

## Regression

```python
# Add a constants
x = sm.add_constant(x1)

# Create a regression with a different method which would be Logit
reg_log = sm.Logit(y,x)

# Fit the regression
result_log = reg_log.fit()

# Regression model
result_log.summary()
```

```python
# Exponential
np.exp(1.9449)
```

# Accuracy

```python
# Predict values
np.set_printoptions(formatter={'float': lambda x: "{0:0.2f}".format(x)})
result_log.predict()
```

```python
# Compare with the actual values
np.array(data['Admitted'])
```

```python
# Summarize using pred table
result_log.pred_table()
```

```python
# Formatting pred table
cm_df = pd.DataFrame(result_log.pred_table())
cm_df.columns = ['Predcit 0', 'Predicted 1']
cm_df = cm_df.rename(index={0:'Actual 0',1:'Actual1'})
cm_df
```

```python
# Calcuulate accuracy
cm = np.array(cm_df)
accuracy_train = (cm[0,0]+cm[1,1])/cm.sum()
accuracy_train
```

```python
# Confusion matrix
def confusion_matrix(data,actual_values,model):
    
    # Make predictions in the giving data
    pred_values = model.predict(data)
    bins=np.array([0,0.5,1])
    
    # Summary the values in a table
    cm = np.histogram2d(actual_values, pred_values, bins=bins)[0]
    
    # Calculate the accuracy
    accuracy = (cm[0,0]+cm[1,1])/cm.sum()
    return cm, accuracy
```

```python
# Test the model
cm = confusion_matrix(test_data,test_actual,result_log)
cm
```
