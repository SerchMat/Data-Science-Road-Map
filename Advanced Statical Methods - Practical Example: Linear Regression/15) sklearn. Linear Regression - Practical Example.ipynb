### Multicollinearity

```python
# Check for multicollinearity
data_cleaned.columns.values
```

```python
# Check for Multicollinearity

# Import VIF model
from statsmodels.stats.outliers_influence import variance_inflation_factor

# Define the feature need for multicollinearity
variables = data_cleaned[['Mileage', 'Year', 'EngineV']]
vif = pd.DataFrame()
vif["VIF"] = [variance_inflation_factor(variables.values, i) for i in range(variables.shape[1])]
vif["features"] = variables.columns
```

```python
# if VIF = 1: no multicollinearity

# Values 1 < VIF < 5 are Ok

# VIF > 6 unacceptable
vif
```

```python
# Since Year has a hig multicollinearity it will be remmove
data_no_multicollinearity = data_cleaned.drop(['Year'],axis=1)
```

## Create dummy variables

```python
# If we have N categorical for a feature, we habe to create N -1 dummines

# Variable to spot all the categorical varibles and create dummines automatically
data_with_dummies = pd.get_dummies(data_no_multicollinearity, drop_first=True)
```

```python
# Check the new data frame
data_with_dummies.head()
```

## Rearrange a bit 

```python
# Check values in the data
data_with_dummies.columns.values
```

```python
# Rearrange the values

#Create a new list
cols = ['log_price','Mileage', 'EngineV', 'Brand_BMW',
       'Brand_Mercedes-Benz', 'Brand_Mitsubishi', 'Brand_Renault',
       'Brand_Toyota', 'Brand_Volkswagen', 'Body_hatch', 'Body_other',
       'Body_sedan', 'Body_vagon', 'Body_van', 'Engine Type_Gas',
       'Engine Type_Other', 'Engine Type_Petrol', 'Registration_yes']
```

```python
# New data frame
data_preprocessed = data_with_dummies[cols]

#Check the values rearranged
data_preprocessed.head()
```

## Linear Regression

### Declare the input and the targets

```python
# Variable to provide the result expected(log price)
targets = data_preprocessed['log_price']

# Inputs
inputs = data_preprocessed.drop(['log_price'],axis=1)
```

### Scale the data

```python
# import necessary libraries
from sklearn.preprocessing import StandardScaler

# Instance of StandardScaler class
scaler = StandardScaler()

# Fit the inputs
scaler.fit(inputs)
```

```python
# Get the satarize inputs
inputs_scaled = scaler.transform(inputs)
```

### Train Test Split

```python
# Split data into training and testing

from sklearn.model_selection import train_test_split

# Variable in which would be split
x_train, x_test, y_train, y_test = train_test_split(inputs_scaled, targets, test_size=0.2, random_state=365)
```

### Create the regression

```python
# Set the variable for Linear regression
reg = LinearRegression()

# Fit linear regresion with the parameters x_train and y_train
reg.fit(x_train, y_train)
```

## Multiple Regression Equation

$\hat{y} = b_0 + b_1x_1 + b_2x_2 + .... + b_kx_k$ 

```python
# Obtain predicted values
y_hat = reg.predict(x_train)
```

```python
# Scatter plt for y-train
plt.scatter(y_train,y_hat)

# Label the data
plt.xlabel('Target(y_train)', size=18)
plt.ylabel('Predictions (y_hat)', size=18)
plt.xlim(6,13)
plt.ylim(6,13)

# Display data
plt.show()
```

```python
# Resisual plot

#Distribution of the residuals
sns.displot(y_train - y_hat)

# Label the data
plt.title('Residuals PDF', size=18)
```

```python
# Calculate R-square
reg.score(x_train, y_train)
```

### Finding the weight and bias

```python
# Reg intercept
reg.intercept_
```

```python
# Reg coefficent
reg.coef_
```

```python
# Summary table with the features and corresponding names
reg_summary = pd.DataFrame(inputs.columns.values, columns=['Features'])
reg_summary['Weights'] = reg.coef_
reg_summary
```

## Testing

```python
# Value store the prediction
y_hat_test = reg.predict(x_test)
```

```python
# Scatter plt for y-test
plt.scatter(y_test,y_hat_test, alpha=0.2) ## alpha specify the opacity (like a heat map)

# Label the data
plt.xlabel('Target(y_test)', size=18)
plt.ylabel('Predictions (y_hat_test)', size=18)
plt.xlim(6,13)
plt.ylim(6,13)

# Display data
plt.show()
```

```python
# Explore the functionality of the algo

# New Data frame
df_pf = pd.DataFrame(np.exp(y_hat_test), columns=['Prediction']) ## np.exp returns exponential of x

# Display the values
df_pf.head()
```

```python
# Compare 
df_pf['Target'] = np.exp(y_test)

# Display the values
df_pf
```
